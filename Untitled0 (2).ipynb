```python
import librosa, librosa.display, numpy as np, matplotlib.pyplot as plt, os

def save_spectrogram(audio_path, save_path):
    try:
        y, sr = librosa.load(audio_path, sr=None)
        S = librosa.feature.melspectrogram(y=y, sr=sr)
        plt.figure(figsize=(2, 2))
        librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
        plt.axis('off')
        plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
        plt.close()
    except Exception as e:
        print(f"âš ï¸ Ø®Ø·Ø£ ÙÙŠ Ù…Ø¹Ø§Ù„Ø¬Ø© Ø§Ù„Ù…Ù„Ù {audio_path}: {e}")

# Ø§Ù„Ù…Ø³Ø§Ø±Ø§Øª Ø§Ù„Ù…ØµØ­Ø­Ø© Ø¨Ù†Ø§Ø¡Ù‹ Ø¹Ù„Ù‰ Ø§Ù„Ù…Ø¬Ù„Ø¯ Ø§Ù„Ø°ÙŠ ÙÙƒÙƒÙ†Ø§ ÙÙŠÙ‡ Ø§Ù„Ø¶ØºØ·
base_mimii_path = '/content/mimii_dataset'

# Ù‚Ù…Ù†Ø§ Ø¨ØªØ¹Ø¯ÙŠÙ„ Ø§Ù„Ù…Ø³Ø§Ø±Ø§Øª Ù„ØªØ·Ø§Ø¨Ù‚ Ø§Ù„Ù‡ÙŠÙƒÙ„Ø© Ø§Ù„Ù…Ø³ØªØ®Ø±Ø¬Ø©
machines = {
    'pump': f'{base_mimii_path}/pump/id_00',
    'fan': f'{base_mimii_path}/fan/id_00',
    'valve': f'{base_mimii_path}/valve/id_00',
    'slider': f'{base_mimii_path}/slider/id_00'
}

print("ğŸš€ Ø¨Ø¯Ø¡ ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ù…Ù„ÙØ§Øª Ø§Ù„ØµÙˆØªÙŠØ© Ø¥Ù„Ù‰ ØµÙˆØ± Spectrograms...")

for m_name, path in machines.items():
    for label in ['normal', 'abnormal']:
        target_dir = os.path.join(path, label)

        # Ø§Ù„ØªØ£ÙƒØ¯ Ù…Ù† ÙˆØ¬ÙˆØ¯ Ø§Ù„Ù…Ø¬Ù„Ø¯ Ù‚Ø¨Ù„ Ø§Ù„Ø¨Ø¯Ø¡
        if os.path.exists(target_dir):
            os.makedirs(f'/content/train_images/{m_name}_{label}', exist_ok=True)
            files = os.listdir(target_dir)[:400]

            for f in files:
                if f.endswith('.wav'): # Ø§Ù„ØªØ£ÙƒØ¯ Ø£Ù†Ù‡ Ù…Ù„Ù ØµÙˆØªÙŠ
                    save_spectrogram(
                        os.path.join(target_dir, f),
                        f'/content/train_images/{m_name}_{label}/{f}.png'
                    )
            print(f"âœ… ØªÙ… ØªØ­ÙˆÙŠÙ„ {m_name} ({label}) Ø¨Ù†Ø¬Ø§Ø­.")
        else:
            print(f"âŒ Ø§Ù„Ù…Ø¬Ù„Ø¯ ØºÙŠØ± Ù…ÙˆØ¬ÙˆØ¯: {target_dir}")

print("\nâœ¨ Ø§ÙƒØªÙ…Ù„Øª Ø§Ù„Ø¹Ù…Ù„ÙŠØ©! Ø§Ù„ØµÙˆØ± Ø§Ù„Ø¢Ù† Ø¬Ø§Ù‡Ø²Ø© ÙÙŠ Ù…Ø¬Ù„Ø¯ /content/train_images")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 1. Ø¥Ø¹Ø¯Ø§Ø¯ Ù…Ø­Ø±Ùƒ Ù‚Ø±Ø§Ø¡Ø© Ø§Ù„ØµÙˆØ± ÙˆØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    '/content/train_images',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    '/content/train_images',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# 2. Ø¨Ù†Ø§Ø¡ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ Ø§Ù„Ø°ÙƒÙŠ (Audio-Image Model)
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5), # Ù„Ø¶Ù…Ø§Ù† Ø¹Ø¯Ù… Ø§Ù„Ø­ÙØ¸ Ø¨Ù„ Ø§Ù„ÙÙ‡Ù…
    layers.Dense(train_generator.num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 3. Ø¨Ø¯Ø¡ Ø§Ù„ØªØ¯Ø±ÙŠØ¨
print("âš™ï¸ Ø¬Ø§Ø±ÙŠ ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ Ø¹Ù„Ù‰ Ø§Ù„Ø¨ØµÙ…Ø§Øª Ø§Ù„ØµÙˆØªÙŠØ©...")
history = model.fit(train_generator, validation_data=val_generator, epochs=10)

# 4. Ø­ÙØ¸ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ Ù„Ø§Ø³ØªØ®Ø¯Ø§Ù…Ù‡ ÙÙŠ Ø§Ù„ÙˆØ§Ø¬Ù‡Ø©
model.save('audio_anomaly_model.h5')
print("âœ… ØªÙ… Ø­ÙØ¸ Ù…ÙˆØ¯ÙŠÙ„ Ø§Ù„Ø¨ØµÙ…Ø© Ø§Ù„ØµÙˆØªÙŠØ© Ø¨Ù†Ø¬Ø§Ø­!")
```
